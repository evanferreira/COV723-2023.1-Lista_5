# Análise dimensional

Seja M, L, T as variáveis fundamentais: massa, comprimento e tempo.

a) Onda de superfície

$$ \omega^2 = kg $$
Sendo $k = \dfrac{2\pi}{\lambda}$ e $g$ a aceleração da gravidade. Temos então: 

$$[\omega] = \dfrac{rad}{s} = 1 \times T^{-1} $$

$$[k] = \dfrac{rad}{m} = 1 \times L^{-1}$$

$$[g] = \dfrac{m}{s^2} = L^1T^{-2}$$

Sendo assim, substituindo na equação, temos: 

$$ [\omega^2] = \dfrac{rad^2}{s^2} = T^{-2} = [kg] = (1 \times L^{-1}) (L^1T^{-2}) = T^{-2} $$

Demonstrando assim, a equivalência dimensional.

b) Onda capilar

$$\omega^2 = \dfrac{k^3\tau}{\rho}$$

Onde $\tau$ é a tensão superifcial (para interface ar-água $\tau=0.07\; N/m$ e $\rho$ é a densidade d'água.

$$[\omega] = \dfrac{rad}{s} = 1 \times T^{-1} $$

$$[k] = \dfrac{rad}{m} = 1 \times L^{-1}$$

$$ [\tau] = kg\dfrac{m}{s^2} \dfrac{1}{m} = M^1L^0T^{-2} $$

$$ [\rho] = \dfrac{kg}{m^3} = M^1L^{-3}$$

$$[\omega^2] = \dfrac{rad^2}{s^2} = T^{-2} = \left[\dfrac{k^3\tau}{\rho} \right] 
= (1 \times L^{-3})(M^1T^{-2})(M^{-1}L^{3}) = T^{-2} $$

Demonstrando assim, a equivalência dimensional. 

c) Onda interna (na interface entre 2 fluidos de densidade diferente)

$$ \omega^2 = kg\left( \dfrac{\rho_2-\rho_1}{\rho_2+\rho_1} \right) $$

A soma ou diferença entre 2 variáveis equivalentes dimensionalmente resulta em um resultado de mesma dimensão, logo os termos $\rho_2-\rho_1$ e $\rho_2+\rho_1$ possuem dimensções iguais vazendo com que a razão $\dfrac{\rho_2-\rho_1}{\rho_2+\rho_1}$ seja adimensional, portanto, analogamente ao item <b>(a)</b> temos a equivalência provada já que: 
    $$ [\omega^2] = \dfrac{rad^2}{s^2} = T^{-2} = [kg]\left[ \dfrac{\rho_2-\rho_1}{\rho_2+\rho_1} \right] = (1 \times L^{-1}) (L^1T^{-2})(1) = T^{-2} $$

# Definições e conceitos
## (a) 
Definimos número de onda como a medida de frequência espacial da onda, ou seja, quantas voltas completas ela dá por unidade de seu comprimento, sendo assim, o número de onda $k$ é definido como: 

$$ k = \dfrac{2\pi}{\lambda}$$

Onde $\lambda$ é o comprimento de onda. 

A frequência angular é uma grandeza análoga, porém, para o domínio temporal, sendo assim, a frequência angular $\omega$ será: 

$$ \omega = \dfrac{2\pi}{T} $$

Onde $T$ é o período de onda.

## (b) 
Porque esta é a velocidade com que a forma da onda e, consequentemente, sua fase se move pelo espaço. Essa grandeza é definida pelo espaço percorrido pelo pico (ou vale) por unidade de tempo, ou seja, a velocidade de fase $c$ é dada pela relção: 

$$c = \dfrac{\lambda}{T} = \dfrac{\omega}{k}$$

## (c) 
A condição de contorno que deve ser respeitada no fundo do canal é a condição de impenetrabilidade, ou seja, a velocidade normal da partícula fluida é nula. 

$$ \dfrac{\partial \vec{V}}{\partial \hat{n}} = \vec{0} $$

Que, considerando a teoria potencial e o sistema de coordenadas cartesiano, ou seja $\vec{V} = \vec{V}(x,y,z) = (\vec{u},\; \vec{v},\; \vec{w})$ , pode ser reescrita da forma: 

$$\vec{w} = \dfrac{\partial \Phi_w }{\partial z} \hat{k} = \vec{0} $$

## (d)
A condição de contorno cinemática define que a velocidade vertical da partícual d'água na superfície liver do fluido é igual à velocidade vertical da superfície em si (condição de não vazamento), sendo a elevação uma função do tempo e distância, temos: 

$$ \dfrac{dz}{dt}\Bigr|_{\substack{z=\zeta}} = \dfrac{D}{Dt}\zeta(x,t)\Bigr|_{\substack{z=\zeta}} 
= \left[\dfrac{\partial\zeta}{\partial t} + \dfrac{\partial\zeta}{\partial x} \dfrac{dx}{dt} \right]_{z=\zeta} 
= \left[\dfrac{\partial\zeta}{\partial t} + u \dfrac{d\zeta}{dx}\right]_{z=\zeta}  $$



## (e) 
A condição de contorno dinâmica vai determinar o equilíbrio de pressões entre a interface onda-ar, ou seja, na superfície livre teremos que a pressão do fluido será equivalente à pressão atmosférica.

## (f) 
Usando a forma geral da condição de contorno cinemática demonstrada no item (c), podemos linearlizar essa relação considerando a hipótese de ondas de pequena declividade, sendo assim teremos

$$u \dfrac{d\zeta}{dx} \to 0$$

Portanto: 

$$\dfrac{dz}{dt}\Bigr|_{\substack{z=\zeta}} = \dfrac{\partial\zeta}{\partial t}\Bigr|_{\substack{z=\zeta}}$$

$$\dfrac{\partial \Phi_w}{\partial z}\Bigr|_{\substack{z=\zeta}} = \dfrac{\partial\zeta}{\partial t}\Bigr|_{\substack{z=\zeta}}$$

Sendo essa a condição de contorno cinemática linearizada.

Partindo do equilíbrio de pressões e com uso da equação de Bernoulli, podemos escrever a condição de contorno dinâmica da seguinte forma:

$p(z=\zeta) = p_0$

$$ \dfrac{\partial \Phi_w}{\partial t} + \dfrac{1}{2} (u^2+v^2+w^2)+\dfrac{p}{\rho} + gz= C$$

Ao linearizar a condição acima e fazendo $z=\zeta$, teremos: 

$$ \dfrac{\partial \Phi_w}{\partial t}+\dfrac{p_0}{\rho} + g\zeta= C$$ 

A constante $p_0/\rho - C$ pode ser incluída dentro da derivada função potencial, portanto, a forma final da condição dinâmica linearizada será: 

$$ \dfrac{\partial \Phi_w}{\partial t}\Bigr|_{\substack{z=\zeta}}+ g\zeta= 0$$ 

Expandindo o potencial em série de Taylor e tendo em mente a hipótese de pequenos deslocamentos verticais, podemos definir essa condição também para $z=0$: 

$$ \dfrac{\partial \Phi_w}{\partial t}\Bigr|_{\substack{z=0}}+ g\zeta= 0$$ 

Como a condição cinemática também é válida também para $z=0$, derivando a condição de contorno dinâmica linearizada acima, podemos escrever: 


$$ \dfrac{\partial^2 \Phi_w}{\partial t^2}\Bigr|_{\substack{z=0}} + g\dfrac{\partial\zeta}{\partial t}\Bigr|_{\substack{z=0}} = 0 $$

$$ \dfrac{\partial z}{\partial t}\Bigr|_{\substack{z=0}} + \dfrac{1}{g}\dfrac{\partial^2 \Phi_w}{\partial t^2}\Bigr|_{\substack{z=0}} = 0 $$

## (g)
Usando a relação da disperção e assumindo a hipótese de águas profundas, temos: 

$$\omega^2 = kg \quad \text{Relação da dispersão para águas profundas}$$

$$ \left(\dfrac{2\pi}{10}\right)^2 = k(9.81) \;\Rightarrow\; \dfrac{2\pi}{\lambda} =  \dfrac{1}{9.81} \left(\dfrac{2\pi}{10}\right)^2 = 0.04 \;rad/m \;\Rightarrow\; \lambda = 157.08 \; m$$

Para $h=1\;m$, faremos a hipótese de águas rasas, sendo assim: 

$$\omega^2 = k^2gh \quad \text{Relação da dispersão para águas rasas}$$

$$ \left(\dfrac{2\pi}{10}\right)^2 = k^2(9.81)(1) \;\Rightarrow\; \dfrac{2\pi}{\lambda} =  \dfrac{1}{\sqrt{9.81}} \dfrac{2\pi}{10} = 0.2 \;rad/m \;\Rightarrow\; \lambda = 31.42 \; m$$

## (h)
Usando a relação da disperção para águas profundas, temos:

$$\omega^2 = kg \quad \text{Relação da dispersão para águas profundas}$$

$$ \left(\dfrac{2\pi}{T}\right)^2 = \dfrac{2\pi}{\lambda}g \;\Rightarrow\; \lambda = \left(\dfrac{T}{2\pi}\right)^2 2\pi g = \dfrac{gT^2}{2\pi} \approx 1.56T^2$$

## (i)
Para águas profundas temos que a velocidade de fase é dada pela relação: 

$$c = \sqrt{\dfrac{g}{k}} = \sqrt{\dfrac{9.81}{0.04}} = 15.66 \; m/s$$

Para águas rasas temos que a velocidade de fase é dada pela relação:

$$ c = \sqrt{gh} = \sqrt{(9.81)(1)} = 3.13 \; m/s $$

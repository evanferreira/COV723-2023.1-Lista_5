# Equação diferencial ordinária de segunda ordem com coeficientes constantes
$\DeclareMathOperator{\sen}{sen}$

$$\ddot{y} + \alpha y = 0$$

Onde $\alpha$ é um número real. 

(a) $\alpha > 0$:

Para $\alpha > 0$, podemos definir $u \;|\; \alpha=u^2$, então temos: 

$$\ddot{y} + u^2 y = 0$$

Sendo assim, o polinômio característico da forma $\lambda^2 + a\lambda+b=0$ será: $\lambda^2 +u^2=0$, portanto teremos 2 raízes distintas e puramente imaginárias da forma $\lambda_{1,2} = \pm ui $, logo a solução geral terá forma: 

$$y(x) = A\sen(ux) + B\cos(ux) =  A\sen(\sqrt{\alpha}x) + B\cos(\sqrt{\alpha}x)$$ 

(b) $\alpha < 0$:

Para $\alpha < 0$, podemos definir $v \;|\; \alpha=-v^2$, então temos: 
$$\ddot{y} - v^2 y = 0$$

Sendo assim, o polinômio característico será: $\lambda^2 -v^2=0$, portanto teremos 2 raízes reais e distintas da forma $\lambda_{1,2} = \pm v$, logo a solução geral terá forma:

$$ y(x) = Ae^{v x} + Be^{-v x} = Ae^{-\sqrt{\alpha} x} + Be^{\sqrt{\alpha} x}$$

(c) $\alpha = 0$:

Nesse caso teremos: 

$$\ddot{y} = 0$$

Integrando 2 vezes, temos: 

$$ y(x) = Ax+B$$

Ao se debruçar sobre a formulação matemática de ondas lineares regulares, temos que lidar com a equação de Laplace $\nabla^2\Phi=0$, recaindo numa EDO de 2º ordem com coeficientes constantes onde $\alpha=0$ e também ao derivar uma expressão para a pressão usando o potencial de velocidades acabamos recaindo numa EDO de 2º ordem do tipo: 

$$ \dfrac{d^2P(z)}{dz^2} - k^2P(z) = 0$$

```python
import numpy as np
from matplotlib import pyplot as plt

x = np.arange(0, 15, 0.1)
alpha_1 = 2
y1 = np.sin(np.sqrt(alpha_1)*x)+0.5*np.cos(np.sqrt(alpha_1)*x)

alpha_2 = -2
y2 = 0.1*np.exp(np.sqrt(-alpha_2*x))+10*np.exp(-np.sqrt(-alpha_2*x))

y3 = 0.5*x+1

fig, ax = plt.subplots(figsize=(12, 5))
ax.plot(x, y1, label=r'$y_1$')
ax.plot(x, y2, label=r'$y_2$')
ax.plot(x, y3, label=r'$y_3$')
ax.grid()
ax.legend()
plt.ylabel('y(x)', rotation=0, labelpad=10)
plt.xlabel('x')
plt.ylim(-3,15)
plt.title('Soluções da EDO')
plt.show()
```

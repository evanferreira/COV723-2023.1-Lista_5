```python
import numpy as np
from matplotlib import pyplot as plt

t = np.linspace(0, 6, 6)
c = 0.25
def f(x): 
    if 0 < x <= 1:
        f=1
    else:
        f=0
    return f
x = np.linspace(-3, 3)
fig, ax = plt.subplots(figsize=(12, 5))

# Item (a):
for T in t:
    f_a = [f(i-c*T) for i in x]
    ax.plot(x, f_a, label=f'f(x-{c*T:.1f})')

ax.grid()
ax.legend()
plt.ylabel('f(x-ct)', rotation=0, labelpad=15)
plt.xlabel('x, m')
plt.title('Forma em movimento - item (a)')
plt.show()

# Item (b):
fig, ax = plt.subplots(figsize=(12, 5))
for T in t:
    f_b = [f(i+c*T) for i in x]
    ax.plot(x, f_b, label=f'f(x+{c*T:.1f})')

ax.grid()
ax.legend()
plt.ylabel('f(x+ct)', rotation=0, labelpad=15)
plt.xlabel('x, m')
plt.title('Forma em movimento - item (b)')
plt.show()
```

# Forma em movimento

Como pôde-se verificar acima, a principal diferença entre as funções (a) e (b) é a direção do movimento, a função $f(x-ct)$ compreende uma progressão no sentido positivo do eixo x, ou seja, para direita, já a função $f(x+ct)$ progride para a esquerda, no sentido negativo do eixo. A seguir os plots das funções (c) e (d).

```python
import numpy as np
from matplotlib import pyplot as plt

t = np.linspace(0, 6, 6)
c = 0.25
def f(x): 
    if 0 < x <= 1:
        f=1
    else:
        f=0
    return f
x = np.linspace(-3, 3)
fig, ax = plt.subplots(figsize=(12, 5))

# Item (c):
for T in t:
    f_c = [f(i-c*T) + f(i+c*T) for i in x]
    ax.plot(x, f_c, label=f'f(x-{c*T:.1f}) + f(x+{c*T:.1f})')

ax.grid()
ax.legend()
plt.ylabel('f(x-ct) + f(x+ct)', rotation=0, labelpad=-2, loc="top")
plt.xlabel('x, m')
plt.title('Forma em movimento - item (c)')
plt.show()

# Item (d):
fig, ax = plt.subplots(figsize=(12, 5))
for T in t:
    f_d = [f(i-c*T) - f(i+c*T) for i in x]
    ax.plot(x, f_d, label=f'f(x-{c*T:.1f}) - f(x+{c*T:.1f})')

ax.grid()
ax.legend()
plt.ylabel('f(x-ct) - f(x+ct)', rotation=0, labelpad=-2, loc="top")
plt.xlabel('x, m')
plt.title('Forma em movimento - item (d)')
plt.show()
```
